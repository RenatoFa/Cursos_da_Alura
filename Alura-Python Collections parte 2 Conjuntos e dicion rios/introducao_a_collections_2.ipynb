type(assistiram)

# Set(Conjunto) // sem repetição de elementos
print(type(set(assistiram))) # não possui indexing do elemento mas é um interavel, exp: assistiram[3]
type(assistiram)

for usuario in assistiram:
    print(usuario)

print(usuarios_data_science | usuarios_machine_learning) # Uniao dos conjuntos sem os elementos repetidos (| elemento ou logico)

print(usuarios_data_science & usuarios_machine_learning) # Interseção dos conjuntos, elementos que estão no conjunto A e B ( & elemento e logico)

print(usuarios_data_science - usuarios_machine_learning) # Diferença dos conjuntos, elementos que estão no conjunto A e não estao no conjunto B ( & elemento e logico)

print(usuarios_data_science ^ usuarios_machine_learning) # Usuarios que estão em A e B mas nao pertencer os dois ao mesmo tempo (^ elemento ou exclusivo)
